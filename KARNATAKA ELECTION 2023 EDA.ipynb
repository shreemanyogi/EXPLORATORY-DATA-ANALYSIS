# INTRODUCTION TO THE ANALYSIS

Karnataka is a state in the southwestern region of India. It was formed as Mysore State on 1 November 1956, with the passage of the States Reorganisation Act, and renamed Karnataka in 1973. The state was part of the Carnatic region in British terminology. Its capital and largest city is Bengaluru.
Legislative Assembly elections were held in Karnataka on 10 May 2023 to elect all 224 members of the Karnataka Legislative Assembly. The votes were counted and the results were declared on 13 May 2023.
Candidates who win the Assembly elections are called 'Member of Lesgislative Assembly' and hold their seats for five years.

# VISUALISATION OF SEATS WON BY EACH PARTY

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='WINNING PARTY', data=ka_data, order=ka_data['WINNING PARTY'].value_counts().index)
plt.title('Number of Seats Won by Each Party in 2023')
plt.xlabel('Party')
plt.ylabel('Number of Seats')
plt.xticks(rotation=45)
plt.show()
```
